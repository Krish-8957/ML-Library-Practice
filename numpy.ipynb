```python
#array attribute
df=np.array([1,2,3,4,5,6])
print(df.ndim)
print(df.shape)#rows and coloumn
print(df.size)
print(df.itemsize)#in bits
print(df.dtype)#use of astype
nf=df.astype(np.int32)
print(nf.dtype)
```

```python
#array operation
df1=np.arange(0,12).reshape(3,4)
df2=np.arange(12,24).reshape(3,4)
print(df1*2)
print(df2==15)
print(df1+df2)
print(np.min(df1))
print(np.max(df1))
print(df1)
print(df2)
```

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame([[1,2],[3,4]])
df2 = pd.DataFrame([[5,6],[7,8]])
print(df1)
a = np.dot(df1, df2)
print(a)
```

```python
#array function
a1 = np.random.random((3, 3))
print(a1)
a1=np.round(a1*100)#return round figurejm
print(a1)
```

```python
nf=np.arange(8).reshape(2,2,2)# 2 dimensional array
print(nf)
```

```python
#indexing and slicing
a1=np.arange(10)
a2=np.arange(12).reshape(3,4)
print(a1)
print(a2)
print(a1[-1])
# slicing
print(a1[2:5])
print(a2[:,0])
print(a2[0,:])
print(a2[1:,1:3])
print(a2[0:2,2:4])
print(a2[0:2,0::3])
```

```python
#for 3d
a3=np.arange(8).reshape(2,2,2)
print(a3)
print(a3[1,0,1])
#if we want 7
print(a3[1,1,1])
print(a3[1,0:1,0:])
```

```python
a4=np.arange(27).reshape(3,3,3)
print(a4)
print(a4[1,0::2,0::2])
print(a4[2,1:3,1:])
```
