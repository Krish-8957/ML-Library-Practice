```python
# plotting a simple graph
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
price=[12000,13000,14000,10000,9000,8000]
revenue=[2018,2019,2020,2021,2022,2023]
plt.plot(revenue,price)
```

```python
# from pandas dataframe
batsman=pd.read_csv('/content/sharma-kohli.csv')
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# draw multiple graph
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('virat kohli vs rohit sharma carrier comaparison')
```

```python
#label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# colour (hex),line(width and style),marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted')
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#line width
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=2)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=3)
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#marker size ,marker= we can change in o,d,in any shape
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=2,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=3,)
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
#leged->location
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=2, label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted',linewidth=3,label='rohit sharma')
plt.title('virat kohli vs rohit sharma carrier comaparison')
plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend( loc ='upper right') # we have to call function calles legend then we can see labeling
```

```python
# limiting axes
price=[12000,13000,14000,10000,9000,8000,440000]
revenue=[2018,2019,2020,2021,2022,2023,2024]
plt.plot(revenue,price)#there is an outlier so the graph of small value see like constant
#if we wan to see graph some accurate we can use
plt.ylim(0,60000)
plt.xlim(2018,2022)
```

```python
#grid if grid not show in other platform then use show()
price=[12000,13000,14000,10000,9000,8000]
revenue=[2018,2019,2020,2021,2022,2023]
plt.plot(revenue,price)
plt.grid()
```
