```python
#series with python functionality
import pandas as pd

marks = [65, 45, 67, 78, 89]
subjects = ['maths', 'english', 'physics', 'computer', 'chemistry']

a = pd.Series(marks, index=subjects)

# Python functions on Series
print("Length:", len(a))
print("Type:", type(a))
print("Max value:", max(a))
print("Min value:", min(a))
print("Sorted values:", sorted(a))
print("Dir (methods):", dir(a))
```

```python
import pandas as pd

a = pd.Series([1, 2, 3])

# int → float
a_float = a.astype(float)

# float → string
a_str = a.astype(str)

print(a_float)
print(a_str)
```

```python
a = pd.Series([10, 20, 30])

print(10 in a.values)   # True
print(40 in a.values)   # False
```

```python
a = pd.Series([1, 2, 3, 4])

for i in a:
    print(i)
```

```python
a = pd.Series([1, 2, 3])
b = pd.Series([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
a = pd.Series([1, 2, 3])
b = pd.Series([2, 2, 2])

print(a > b)
print(a < b)
print(a == b)
print(a != b)
```
